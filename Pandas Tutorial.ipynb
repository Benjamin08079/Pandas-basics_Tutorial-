# Series create, manipulate, query, delete

```python
# creating a series from a list.
arr = [0,1,2,3,4]
s1 = pd.Series(arr)
s1
```

```python
order = [1,2,3,4,5]
s2 = pd.Series(arr, index=order)
s2
```

```python
import numpy as np
n=np.random.randn(5) # Creat a random Ndarray
index = ['a','b','c','d','e']
s2 = pd.Series(n, index=index)
s2
```

```python
# create series from dictionary
d={'a':1,'b':2,'c':3,'d':4,'e':5}
s3=pd.Series(d)
s3
```

```python
# you can modify the index of series
print(s1)
s1.index=['A','B','C','D','E']
s1
```

# series operations

```python
arr1=[0,1,2,3,4,5,7]
arr2=[6,7,8,9,5]
```

```python
s5=pd.Series(arr2)
s5
```

```python
s6 = pd.Series(arr1)
s6
```

```python
s5.add(s6)
```

```python
s5.sub(s6)
```

```python
s7=s5.mul(s6)
s7
```

```python
s5.div(s6)
```

```python
print('median', s7.median())
print('max', s7.max())
print('min', s7.min())
```
